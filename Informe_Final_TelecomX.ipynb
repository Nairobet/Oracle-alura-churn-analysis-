#**INFORME FINAL – Análisis de Evasión de Clientes (Churn) en Telecom X**

##**1. Resumen Ejecutivo**

Telecom X enfrenta una tasa de abandono del 26.6%, lo que significa que 1 de cada 4 clientes cancela el servicio. Este nivel es considerado alto dentro de la industria de telecomunicaciones y justifica la necesidad de un análisis profundo para identificar los factores que impulsan el churn.
El análisis exploratorio realizado permitió identificar cinco drivers principales que explican la mayor parte del abandono:
- Tipo de contrato
- Antigüedad del cliente
- Cargo mensual
- Método de pago
- Perfil familiar
Los resultados muestran patrones claros y accionables: los clientes con contratos mensuales, recién ingresados, con cargos altos, que pagan mediante métodos no automáticos y que viven solos presentan las tasas más elevadas de abandono.
Estos hallazgos permiten orientar estrategias de retención focalizadas y sirven como base para un futuro modelo predictivo de churn.




##**2. Objetivo del Proyecto**

El propósito de este análisis es:
- Explorar, limpiar y preparar los datos entregados por Telecom X.
- Identificar patrones y factores asociados al abandono de clientes.
- Generar insights accionables para el equipo de negocio y el equipo de ciencia de datos.
- Establecer una base sólida para un modelo predictivo de churn.


##**3. Metodología**

###**3.1 Extracción**

Los datos fueron obtenidos desde un archivo JSON estructurado en múltiples niveles. Se utilizó pd.json_normalize para expandir las secciones:
- customer
- phone
- internet
- account
Luego se consolidaron en un único DataFrame.


###**3.2 Transformación**

Se realizaron las siguientes tareas:
- Conversión de Charges.Total a numérico.
- Eliminación de registros con valores nulos.
- Limpieza de la columna Churn (espacios, valores vacíos).
- Renombrado de columnas a español para facilitar la interpretación.
- Conversión de variables categóricas “Yes/No” a 0/1.
- Creación de nuevas variables derivadas (grupos de antigüedad, grupos de cargo mensual, perfil familiar)


###**3.3 Análisis Exploratorio (EDA)**

Se aplicaron comparativas entre churn y variables clave:
- Tipo de contrato
- Antigüedad
- Cargo mensual
- Método de pago
- Perfil familiar
Cada análisis incluyó:
- Tablas porcentuales normalizadas
- Tablas estilizadas
- Gráficos personalizados
- Interpretación detallada
- Insights accionables




##**4. Hallazgos Clave por Variable**

###**4.1 Tipo de contrato**

- Los clientes con contrato Month-to-Month presentan la mayor tasa de abandono (≈43%).
- Los contratos anuales o de dos años reducen el churn drásticamente.
Insight:
La ausencia de compromiso facilita la cancelación. Este segmento debe ser prioridad en retención.





contrato.png

###**4.2 Antigüedad del cliente**

- Los clientes con 0–6 meses presentan un churn crítico de 53%.
- A mayor antigüedad, menor probabilidad de abandono.
Insight:
El onboarding es el punto más vulnerable del ciclo de vida del cliente.





antiguedad.png

###**4.3 Cargo mensual**

- Los clientes con cargos altos o muy altos tienen tasas de abandono de hasta 38%.
- Los clientes con cargos bajos presentan solo 11%.
Insight:
La sensibilidad al precio es un driver fuerte. Planes altos requieren beneficios adicionales o descuentos.





cargo.png

###**4.4 Método de pago**

- El método Electronic Check presenta la tasa más alta (45%).
- Métodos automáticos (tarjeta, transferencia bancaria) reducen el churn.
Insight:
Los métodos no automáticos generan fricción y aumentan la probabilidad de cancelación.





metodo.pago.png

###**4.5 Perfil familiar**

- El segmento con mayor riesgo es Adulto mayor viviendo solo (≈49%).
- Los clientes con pareja y dependientes presentan el churn más bajo.
Insight:
La estructura familiar influye en la dependencia del servicio y en la sensibilidad al precio.

##**6. Conclusiones**

- El churn en Telecom X no es aleatorio: está fuertemente asociado a condiciones específicas del cliente y del servicio.
- Existen segmentos claramente identificables con riesgo elevado.
- El análisis exploratorio permitió transformar datos operativos en información accionable.
- Los resultados son consistentes y ofrecen una base sólida para un modelo predictivo de churn.




##**7. Recomendaciones Estratégicas**

###**7.1 Retención temprana (primeros 6 meses)**

- Programas de bienvenida.
- Seguimiento proactivo.
- Beneficios exclusivos para nuevos clientes


###**7.2 Migración de contratos**

- Incentivar el paso de Month-to-Month a contratos anuales.
- Ofrecer descuentos por permanencia.


###**7.3 Optimización de precios**

- Revisar planes altos.
- Ofrecer bundles o beneficios adicionales.


###**7.4 Automatización de pagos**

- Incentivar métodos automáticos.
- Descuentos por domiciliación bancaria.


###**7.5 Segmento vulnerable: adultos mayores solos**

- Programas de soporte especializado.
- Atención personalizada.
- Simplificación de procesos
